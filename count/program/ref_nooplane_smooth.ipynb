```python
import os
import ijson
import json
```

```python
        
# count a REF data file 
def count_ref_onefile(filename):
    result={}
    num_total=[0,0,0]
    num_nofix=[0,0,0]
    num_o9=[0,0,0]
    result.setdefault("num_total:[ref1, ref2, ref3, sumall]", num_total)
    result.setdefault("num_nofix:[ref1, ref2, ref3, sumall]", num_nofix)
    result.setdefault("num_o9:[ref1, ref2, ref3, sumall]", num_o9)
    with open(filename, 'r') as p:
        datas=ijson.items(p, 'item')
        for data in datas:
            if data=={} or data==[]:
                continue
            for ref in [0,1,2]:
                if "invol"+str(ref+1) in data.keys():
                    for dataone in data["invol"+str(ref+1)]:
                        if "True]" in dataone:
                            if "O9" in dataone:
                                num_o9[ref]+=1
                            elif "total" in dataone:
                                num_total[ref]+=1
                            elif "No fixed" in dataone:
                                num_nofix[ref]+=1
    num_total.append(sum(num_total))
    num_nofix.append(sum(num_nofix))
    num_o9.append(sum(num_o9))
    result["num_total:[ref1, ref2, ref3, sumall]"]= num_total
    result["num_nofix:[ref1, ref2, ref3, sumall]"]= num_nofix
    result["num_o9:[ref1, ref2, ref3, sumall]"] = num_o9
    return result


# only smooth(h21+-  integer)
def count_nooplane_smooth_many_files(filepath):
    result={}
    fileori=os.listdir(filepath)
    num_total=[0,0,0,0]
    num_nofix=[0,0,0,0]
    num_o9=[0,0,0,0]
    result.setdefault("num_total:[ref1, ref2, ref3, sumall]", num_total)
    result.setdefault("num_nofix:[ref1, ref2, ref3, sumall]", num_nofix)
    result.setdefault("num_o9:[ref1, ref2, ref3, sumall]", num_o9)
    for file in fileori:
        try:            
            if os.path.isdir(filepath+file):
                resultonefile=count_nooplane_smooth_many_files(filepath+file+"/")
            else:
                resultonefile=count_ref_onefile(filepath+file)
        except:
            print(filepath+file)
            continue
        for key in result.keys():
            if type(result[key])==list:
                for i in range(len(result[key])):
                    result[key][i]+=resultonefile[key][i]
            else:
                print("这个key是什么类型的，不太清楚")
    return result


```

```python
file2to7="../../dataout/REF/"
```

```python
filepaths=os.listdir(file2to7)
datasout=[]
for filepath in filepaths:
    if os.path.isdir(file2to7+filepath):
        dataout=count_nooplane_smooth_many_files(file2to7+filepath+"/")
        dataout.setdefault("file", filepath)    
        datasout.append(dataout)
with open("../result/2-7refnooplane_smooth.json", 'w') as pout:
    json.dump(datasout, pout, indent=2)
```
